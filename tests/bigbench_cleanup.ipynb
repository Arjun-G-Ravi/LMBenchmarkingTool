```python
# RAM killer
import os
import json
import csv
from pathlib import Path

def extract_examples_to_csv(directory, output_file):
    examples_data = []
    
    for root, _, files in os.walk(directory):
        if 'task.json' in files:
            json_path = Path(root) / 'task.json'
            try:
                with open(json_path) as f:
                    data = json.load(f)
                    if 'name' in data and 'examples' in data:
                        task_name = data['name']
                        for example in data['examples']:
                            examples_data.append({
                                'task_name': task_name,
                                'input': example.get('input', ''),
                                'target': example.get('target', '')
                            })
            except json.JSONDecodeError:
                print(f"Error: Invalid JSON in {json_path}")
            except KeyError as e:
                print(f"Error: Missing required field {e} in {json_path}")

    if examples_data:
        with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
            fieldnames = ['task_name', 'input', 'target']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(examples_data)
        print(f"Successfully wrote {len(examples_data)} examples to {output_file}")
    else:
        print("No examples found in any task.json files")

if __name__ == "__main__":
    directory = "/home/arjun/Desktop/GitHub/LMBenchmarkingTool/data/bigbench"
    output_file = "task_examples.csv"
    
    if not os.path.isdir(directory):
        print(f"Error: '{directory}' is not a valid directory")
    else:
        extract_examples_to_csv(directory, output_file)
```

```python
import os
import json
import csv
from pathlib import Path

def process_json_file(json_path, writer):
    try:
        with open(json_path) as f:
            data = json.load(f)
            if 'name' in data and 'examples' in data:
                task_name = data['name']
                for example in data['examples']:
                    writer.writerow({
                        'task_name': task_name,
                        'input': example.get('input', ''),
                        'target': example.get('target', '')
                    })
                return 1  # Return count of processed files
    except json.JSONDecodeError:
        print(f"Error: Invalid JSON in {json_path}")
    except KeyError as e:
        print(f"Error: Missing required field {e} in {json_path}")
    return 0

def extract_examples_to_csv(directory, output_file, batch_size=100):
    processed_files = 0
    processed_examples = 0
    
    with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['task_name', 'input', 'target']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        
        for root, _, files in os.walk(directory):
            if 'task.json' in files:
                json_path = Path(root) / 'task.json'
                files_processed = process_json_file(json_path, writer)
                processed_files += files_processed
                
                # Print progress periodically
                if processed_files % batch_size == 0:
                    print(f"Processed {processed_files} files...")

    print(f"Completed! Processed {processed_files} files in total")

if __name__ == "__main__":
    directory = "/home/arjun/Desktop/GitHub/LMBenchmarkingTool/data/bigbench"
    output_file = "task_examples.csv"
    
    if not os.path.isdir(directory):
        print(f"Error: '{directory}' is not a valid directory")
    else:
        extract_examples_to_csv(directory, output_file)
```
